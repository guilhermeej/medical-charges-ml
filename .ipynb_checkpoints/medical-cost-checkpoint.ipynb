```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import plotly.express as px
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('datasets/insurance.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.histplot(df, x='age')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.histplot(df, x='age', hue='smoker', ax=ax)
ax.set_xlabel('Idade')
ax.set_ylabel('Quantidade')
ax.set_title('Distribuio de Idade Sendo Fumante')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.histplot(df, x='bmi', ax=ax)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(df, x='age', y='bmi',hue='smoker', ax=ax)
ax.set_title('Idade vs IMC')
ax.set_xlabel('Idade')
ax.set_ylabel('IMC')
plt.show()
```

```python
print(f'Mdia de IMC dos fumantes: {df[df['smoker'] == 'yes']['bmi'].mean():.2f}')
print(f'Mdia de IMC dos no fumantes: {df[df['smoker'] == 'no']['bmi'].mean():.2f}')
```

```python
print(f'Desvio padro de IMC dos fumantes: {df[df['smoker'] == 'yes']['bmi'].std():.2f}')
print(f'Desvio padro de IMC dos no fumantes: {df[df['smoker'] == 'no']['bmi'].std():.2f}')
```

```python
sns.displot(df, x='region')
plt.show()
```
